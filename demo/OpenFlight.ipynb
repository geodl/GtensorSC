```julia
# demo for openFlight data

# para is the parameters for the algorithm here we set ALPHA = 0.8, MIN_NUM = 5, MAX_NUM = 100, PHI = 0.4
para = algPara(0.8, 5, 100,0.4)
P = read_tensor("../data/openFlight/tensor_openflight.txt");

# the return value r is the binary tree for clustering results.
(r,h) = tensor_speclustering(P, para);

# after the clustering algorithm completed, to output the clustering result
# use the code: indVec = zeros(Int,r.n); traCount = trav_tree(r, indVec, 1)
# indVec then shows the group number for each node as indexed by line number
```

```julia
# load the openflight dic
f = readdlm("../data/openFlight/dic_openflight.txt");
wordDic = [f[i,1] => string(int(f[i,2])) for i = 1:size(f,1)];

airlineDic = Dict();
tt = open("../data/openFlight/airline.txt","r")
ss = readline(tt);
while ss!=""
    nn = sizeof(ss)
    i = 1
    while ss[i]!='#'
        i = i+1
    end
    id = ss[1:i-1]
    dec = ss[i+1:nn-1]
    airlineDic[id]=dec
    ss = readline(tt)
end


airportDic = Dict();
tt = open("../data/openFlight/airport.txt","r")
ss = readline(tt);
while ss!=""
    nn = sizeof(ss)
    i = 1
    while ss[i]!='#'
        i = i+1
    end
    id = ss[1:i-1]
    dec = ss[i+1:nn-1]
    airportDic[id]=dec
    ss = readline(tt)
end
```

```julia
# indVec is the clustering result indicating group number for each node
indVec = zeros(r.n)
traCount = trav_tree(r, indVec, 1)

# print groups from start_num to end_num
# This is the europe cluster
start_num = 113; end_num = 125
countLine = 0; countPort = 0;
for i=start_num:end_num
    for j=1:539
        if indVec[j] == i
            countLine+=1
            println(airlineDic[string(wordDic[j])])
        end
    end
    for j=540:3478
        if indVec[j]==i
            countPort+=1
            println(airportDic[string(wordDic[j])])
        end
    end
end
println("Total airline is: $(countLine). Total airport is: $(countPort)")
```

```julia
# This is the US cluster
start_num = 108; end_num = 112
countLine = 0; countPort = 0;
for i=start_num:end_num
    for j=1:539
        if indVec[j] == i
            countLine+=1
            println(airlineDic[string(wordDic[j])])
        end
    end
    for j=540:3478
        if indVec[j]==i
            countPort+=1
            println(airportDic[string(wordDic[j])])
        end
    end
end
println("Total airline is: $(countLine). Total airport is: $(countPort)")
```

```julia
# This script generate the output data for matlab plot
f = open("../data/openFlight/result_airline.txt","w")
for i=1:traCount-1
    for j=1:539
        if indVec[j]==i
            write(f,"$(j)\n")
        end
    end
end
close(f)


f = open("../data/openFlight/result_airport.txt","w")
for i=1:traCount-1
    for j=540:r.n
        if indVec[j]==i
            write(f,"$(j)\n")
        end
    end
end
close(f)
```
